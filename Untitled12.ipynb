```
import pydicom as dicom
import matplotlib.pylab as plt
import os
import cv2
import numpy as np
# specify your image path
#-1-
path = '/content/01/DICOM/PA000001/ST000000/SE000009/IM000553'
# path = '/content/01/DICOM/PA000000/ST000000/SE000002/IM000026'

ds = dicom.dcmread(path)

plt.figure(figsize=(15,15))
plt.imshow(ds.pixel_array, cmap=plt.cm.bone)
plt.show()
if 'WindowWidth' in ds:
    print('Dataset has windowing')
```

```
image_2d = ds.pixel_array.astype(float)

# Rescaling grey scale between 0-255
image_2d_scaled = (np.maximum(image_2d,0) / image_2d.max()) * 255.0

# Convert to uint
image_2d_scaled = np.uint8(image_2d_scaled)

img = image_2d_scaled
cv2.imwrite('image.png',img)
```

```
# cv2_imshow(img)
```

```
ret,threshOUT = cv2.threshold(img,100,128,cv2.THRESH_BINARY)
ret,threshIN = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
#-------------------------

kernel = np.ones((5,5),np.uint8)
threshOUT = cv2.morphologyEx(threshOUT, cv2.MORPH_OPEN, kernel)
#---------------------------------
# cv2_imshow(threshOUT)
# cv2_imshow(threshIN)
```

```
image = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)

start_point = (205, 230) 
end_point = (280, 293) 

color = (0, 0, 255) 
   
thickness = 1
image = cv2.rectangle(image, start_point, end_point, color, thickness) 
#------somewhere there is an edge-----------
cv2_imshow(image)
# cv2_imshow(img)
```

```
image_with_area = threshOUT[start_point[1]:end_point[1], start_point[0]:end_point[0]]
cv2_imshow(image_with_area)
pix_sum = np.sum(image_with_area)/128
pix_num = image_with_area.shape[0]*image_with_area.shape[1]
#mean value
mu = pix_sum/pix_num
#standard deviation
var = 0
for i in range(0, image_with_area.shape[0]):
  for j in range(0, image_with_area.shape[1]):
    var += ((image_with_area[i,j]/128 - mu)**2)/pix_num
std = np.sqrt(var)
print('the mean is', mu, 'the standard deviation is', std, 'the variance is', var)
```

```
#into pdf
pdf = np.zeros([image_with_area.shape[0],image_with_area.shape[1]])
for i in range(0, image_with_area.shape[0]):
  for j in range(0, image_with_area.shape[1]):
    pdf[i,j] = 1/(np.sqrt(2*var*np.pi))*np.exp((-(image_with_area[i,j]/128 - mu)**2)/2*var)
weights = ((pdf-np.min(pdf))/(np.max(pdf)-np.min(pdf)))*100
```

```
image_to_draw = cv2.cvtColor(image_with_area, cv2.COLOR_GRAY2RGB)
table = np.zeros([image_with_area.shape[0],image_with_area.shape[1]])
for i in range(0, image_with_area.shape[0]-4):
  for j in range(0, image_with_area.shape[1]-4):
    table[i,j] = np.sum((image_with_area[i:i+4,j:j+4]/128)*weights[i,j]/4)
    if table[i,j] > 50:
      image_to_draw[i,j] = (0,0,225)
cv2_imshow(image_to_draw)
#this is the segmented part we need
```

```
b,g,r = cv2.split(image_to_draw)
contour_img = r
cv2_imshow(contour_img)

ret, thresh = cv2.threshold(contour_img, 220, 255, 0)
cv2_imshow(thresh)
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

#find the biggest contour
largest_area=0
for i in contours:
    area = cv2.contourArea(i)
    if (area>largest_area):
        largest_area=area
        largest_contour_index=i

contour_img = cv2.cvtColor(contour_img, cv2.COLOR_GRAY2RGB)
cnt = largest_contour_index
cv2.drawContours(contour_img, [cnt], 0, (0,255,0), 1)
cv2_imshow(contour_img)
#this is the contoured image
```

```
def find_contour (contour_img):
  #dilate contour
  b,g,r = cv2.split(contour_img)
  ret, Gthresh = cv2.threshold(g, 250, 255, 0)
  kernel = np.ones((8,8),np.uint8)
  dilated_contour_img = cv2.dilate(Gthresh,kernel,iterations = 1)
  GCBACcontours, hierarchy = cv2.findContours(dilated_contour_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
  #smooth contour
  from scipy.interpolate import splprep, splev
  new_contour = cv2.cvtColor(dilated_contour_img, cv2.COLOR_GRAY2RGB)
  smoothened = []
  for contour in GCBACcontours:
      x,y = contour.T
      # Convert from numpy arrays to normal arrays
      x = x.tolist()[0]
      y = y.tolist()[0]
      # https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.interpolate.splprep.html
      tck, u = splprep([x,y], u=None, s=1.0, per=1)
      # https://docs.scipy.org/doc/numpy-1.10.1/reference/generated/numpy.linspace.html
      u_new = np.linspace(u.min(), u.max(), 25)
      # https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.interpolate.splev.html
      x_new, y_new = splev(u_new, tck, der=0)
      # Convert it back to numpy format for opencv to be able to display it
      res_array = [[[int(i[0]), int(i[1])]] for i in zip(x_new,y_new)]
      smoothened.append(np.asarray(res_array, dtype=np.int32))

  # Overlay the smoothed contours on the original image
  smallest_area=10000000
  for i in smoothened:
      area = cv2.contourArea(i)
      if (area<smallest_area):
          smallest_area=area
          smallest_contour_index=i

  cnt = smallest_contour_index
  cv2.drawContours(new_contour, [cnt], 0, (0,0,225), 1)

  b,g,r = cv2.split(new_contour)
  r = r-b-g
  ret, thresh_r = cv2.threshold(r, 220, 255, 0)

  #again dilating and taking OUTER contour
  dilated_r = cv2.dilate(thresh_r,kernel,iterations = 1)
  contours, hierarchy = cv2.findContours(dilated_r, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
  largest_area=0
  for i in contours:
      area = cv2.contourArea(i)
      if (area>largest_area):
          largest_area=area
          largest_contour_index=i

  new_r = cv2.cvtColor(dilated_r, cv2.COLOR_GRAY2RGB)
  cnt = largest_contour_index
  cv2.drawContours(new_r, [cnt], 0, (0,0,255), 1)
  length = cv2.arcLength(cnt, True)

  b,g,r = cv2.split(new_r)
  red = r-b-g
  ret, red = cv2.threshold(red, 250, 255, 0)

  return(red, length)

red, length = find_contour(contour_img)
cv2_imshow(red)
print('length of contour is', length)
```
